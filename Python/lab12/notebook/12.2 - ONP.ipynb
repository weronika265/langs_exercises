Proszę napisać funkcję, która obliczy wynik działania podanego za pomocą ONP.

[Opis ONP na Wikipedii](https://pl.wikipedia.org/wiki/Odwrotna_notacja_polska)

Przyjmijmy tylko podstawowe operacje: dodawanie, odejmowanie, mnożenie, dzielenie. Poszczególne symbole są rozdzielane spacjami.

# Algorytm:

Dla wszystkich symboli z wyrażenia ONP wykonuj:
- jeśli i-ty symbol jest liczbą, to odłóż go na stos,
- jeśli i-ty symbol jest operatorem to:
  - zdejmij ze stosu jeden element (ozn. a),
  - zdejmij ze stosu kolejny element (ozn. b),
  - odłóż na stos wartość b operator a.
- Zdejmij ze stosu wynik.

````python
def onp(str):
    ''' Funkcja oblicza ciąg podanych wyrażeń jako Odwrotną Notację Polską.
        Zabiera na stos liczby i kiedy napotka znak, zabiera ze stosu dwie ostatnie liczby. 
        Wykonuje na nich działania, a ten wynik zwraca na stos.
        W trakcie wypisuje działania na wyraażeniu.
        Zwraca wynik działania. '''
    wyrazenie = []
    for l in str.split():
        wyrazenie.append(l)
        
    stos = []
    while wyrazenie != []:
        print('wyrażenie: %s' % wyrazenie)
        print('stos: %s' % stos)
        val = wyrazenie.pop(0)
        if val in ['+', '-', '*', '/']:
            a = stos.pop()
            b = stos.pop()
            if val == '+':
                wynik = b + a
            elif val == '-':
                wynik = b - a
            elif val == '*':
                wynik = b * a
            elif val == '/':
                wynik = b / a
            stos.append(wynik)
            print('zdjęty element: "%s" wynik operacji: "%s%s%s" trafia na stos\n' % (val, b, val, a))
        else:
            stos.append(int(val))
            print('zdjęty element: "%d" trafia na stos\n' % int(val))
    return stos.pop()

print( onp('2 3 + 5 *') )
print('\n-----------------------------------\n')
print( onp('5 7 + 3 / 12 6 - 4 * + 2 /') )
````
